```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import time
```

```python
data = pd.read_excel("C:/Users/HoonCheol/workspace/New_Tokenizer/토큰화/token1.xlsx")
token = data["hannanum_token"] # 한나눔 명사 토큰들
questions = [] # 복합명사 리스트
counts = [] # 검색결과 수 리스트
```

```python
#드라이버 경로 지정
path = 'c:/Users/HoonCheol/chromedriver.exe' ## 크롬드라이브가 설치된 경로 지정
driver = webdriver.Chrome(path)
basic_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query='
```

# 1. 2개 붙인 복합명사 만들고 갯수로 판단

```python
for i in token:
    list_token = i.split() # 한 행의 명사 토큰들
    for j, noun in enumerate(list_token[:-1]):
        question = noun + list_token[j+1] #복합명사 만들기
        driver.get(basic_url+question)
        delay = 0.1
        driver.implicitly_wait(delay)
        
        search = driver.find_element_by_class_name("main_pack")
        
        #한글만
        onlyKorean = re.compile("[^ㄱ-ㅣ가-힣\s]+")
        korean = onlyKorean.sub('',search.get_attribute("innerHTML"))
        
        #검색 결과 존재하면 복합명사로 판단
        if korean.count(question) - 11 > 0:
            questions.append(question)
            counts.append(korean.count(question) - 11)
```

```python
multi_nouns = pd.DataFrame({"complex" : questions, "count": counts})
```

```python
multi_nouns.to_csv("complex_1_hannanum_2.csv", index = False, encoding = 'utf-8-sig')
```
